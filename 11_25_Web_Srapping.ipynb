```python
import re
import urllib
import pandas as pd
```

```python
from urllib.request import urlopen
url="https://www.summet.com/dmsi/html/codesamples/addresses.html"
text=urlopen(url).read().decode()
#print(text)

contact_number=re.findall('\(\d{3}\)\s\d{3}[-]\d{4}',text)
#print("contact number= ",contact_number)


zip_code=re.findall(r'\b\d{5}',text)
#print("zip code= ",zip_code)

data={"contact number":contact_number,
    "zip cod":zip_code}
print(data)

# df=pd.DataFrame(data)
# df
print(len(contact_number))
print(len(zip_code))

```
